```python
# Import thư viện
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
# Đọc dữ liệu
df = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head(10)
```

```python
# Xử lý data
df = df.drop(columns='customerID')
df.head(10)
```

#Chuẩn hóa, làm phẳng dữ liệu


```python
# Xử lý biến phụ thuộc
def process_churn(churn):
    if churn == 'Yes':
        return 1
    else:
        return 0
df['Churn'] = df['Churn'].apply(process_churn)
```

```python
# Hàm xử lý gender
def process_gender(gender):
    if gender == 'Male':
        return 1
    else:
        return 0
df['gender'] = df['gender'].apply(process_gender)
```

```python
# Hàm xử lý Partner
def process_partner(partner):
    if partner == 'Yes':
        return 1
    else:
        return 0
df['Partner'] = df['Partner'].apply(process_partner)
```

```python
# Hàm xử lý Dependents
def process_dependents(dependents):
    if dependents == 'Yes':
        return 1
    else:
        return 0
df['Dependents'] = df['Dependents'].apply(process_dependents)
```

```python
# Hàm xử lý PhoneService
def process_phoneservice(phoneservice):
    if phoneservice == 'Yes':
        return 1
    else:
        return 0
df['PhoneService'] = df['PhoneService'].apply(process_phoneservice)
```

```python
# Hàm xử lý MultipleLines
def process_multilines(multilines):
    mapping = {'No phone service': 2, 'No': 0 , 'Yes': 1}
    return mapping.get(multilines)
df['MultipleLines'] = df['MultipleLines'].apply(process_multilines)
```

```python
# Hàm xử lý InternetService
def process_internetservice(internetservice):
    mapping = {'No': 0, 'DSL': 1 , 'Fiber optic': 2}
    return mapping.get(internetservice)
df['InternetService'] = df['InternetService'].apply(process_internetservice)
```

```python
# Hàm xử lý OnlineSecurity
def process_onlinesecurity(onlinesecurity):
    mapping = {'No internet service': 2, 'No': 0 , 'Yes': 1}
    return mapping.get(onlinesecurity)
df['OnlineSecurity'] = df['OnlineSecurity'].apply(process_onlinesecurity)
```

```python
# Hàm xử lý OnlineBackup
def process_onlinebackup(onlinebackup):
    mapping = {'No internet service': 2, 'No': 0 , 'Yes': 1}
    return mapping.get(onlinebackup)
df['OnlineBackup'] = df['OnlineBackup'].apply(process_onlinebackup)
```

```python
# Hàm xử lý TechSupport
def process_techsupport(techsupport):
    mapping = {'No internet service': 2, 'No': 0 , 'Yes': 1}
    return mapping.get(techsupport)
df['TechSupport'] = df['TechSupport'].apply(process_techsupport)
```

```python
# Hàm xử lý DeviceProtection
def process_deviceprotection(deviceprotection):
    mapping = {'No internet service': 2, 'No': 0 , 'Yes': 1}
    return mapping.get(deviceprotection)
df['DeviceProtection'] = df['DeviceProtection'].apply(process_deviceprotection)
```

```python
# Hàm xử lý StreamingTV
def process_streamingtv(streamingtv):
    mapping = {'No internet service': 2, 'No': 0 , 'Yes': 1}
    return mapping.get(streamingtv)
df['StreamingTV'] = df['StreamingTV'].apply(process_streamingtv)
```

```python
# Hàm xử lý StreamingMovies
def process_streamingmovies(streamingmovies):
    mapping = {'No internet service': 2, 'No': 0 , 'Yes': 1}
    return mapping.get(streamingmovies)
df['StreamingMovies'] = df['StreamingMovies'].apply(process_streamingmovies)
```

```python
# Hàm xử lý Contract
def process_contract(contract):
    mapping = {'Month-to-month': 0, 'One year': 1 , 'Two year': 2}
    return mapping.get(contract)
df['Contract'] = df['Contract'].apply(process_contract)
```

```python
# Hàm xử lý PaperlessBilling
def process_paperlessbilling(paperlessbilling):
    if paperlessbilling == 'Yes':
        return 1
    else:
        return 0
df['PaperlessBilling'] = df['PaperlessBilling'].apply(process_paperlessbilling)
```

```python
# Hàm xử lý PaymentMethod
def process_paymentmethod(paymentmethod):
    mapping = {'Electronic check': 0, 'Mailed check': 1 , 'Bank transfer (automatic)': 2, 'Credit card (automatic)': 3}
    return mapping.get(paymentmethod)
df['PaymentMethod'] = df['PaymentMethod'].apply(process_paymentmethod)
```

```python
# Chuyển data type TotalCharges
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

# Train Model


```python
# Import thư viện chia tệp và làm phẳng dữ liệu
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Chọn biến phụ thuộc và biến độc lập
y = df['Churn']
x = df.drop(columns='Churn')
```

```python
# Chia tập dữ liệu thành tập huấn luyện và tập kiểm tra (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Tiền xử lý dữ liệu: chuẩn hóa dữ liệu
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Import các chỉ số đánh giá máy học
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
```

```python
# Import thư viện BiLSTM - CNN
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, Bidirectional, Flatten, Dropout
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
# Định hình lại dữ liệu để phù hợp với đầu vào của LSTM và CNN (dạng 3D: mẫu, bước thời gian, đặc trưng)
x_train_reshaped = x_train_scaled.reshape(x_train_scaled.shape[0], 1, x_train_scaled.shape[1])
x_test_reshaped = x_test_scaled.reshape(x_test_scaled.shape[0], 1, x_test_scaled.shape[1])

# Xây dựng mô hình BiLSTM-CNN
model = Sequential()

# Lớp BiLSTM với 64 units
model.add(Bidirectional(LSTM(64, return_sequences=True), input_shape=(1, x_train_scaled.shape[1])))

# Lớp CNN (Conv1D) với 64 bộ lọc, kernel size = 1
model.add(Conv1D(64, kernel_size=1, activation='relu'))

# Lớp MaxPooling1D để giảm số lượng đặc trưng
model.add(MaxPooling1D(pool_size=1))

# Lớp Flatten để biến đổi dữ liệu thành 2D
model.add(Flatten())

# Lớp Dense với 64 neuron và regularization l2 để giảm overfitting
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.01)))

# Lớp Dropout với 0.8 để giảm overfitting
model.add(Dropout(0.8))

# Lớp Dense cuối cùng để phân loại nhị phân với hàm sigmoid
model.add(Dense(1, activation='sigmoid'))

# Compile mô hình
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Huấn luyện mô hình
model.fit(x_train_reshaped, y_train, batch_size=32, epochs=20, validation_split=0.2)

# Dự đoán trên tập kiểm tra
y_pred = (model.predict(x_test_reshaped) > 0.5).astype("int32")

# Tính toán các chỉ số đánh giá
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# In ra kết quả
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-score: {f1:.4f}')
```
